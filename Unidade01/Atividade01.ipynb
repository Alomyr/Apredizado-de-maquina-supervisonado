# 2.1 Dados os vetores a = [1, 2, 3] e b = [4, 5, 6], calcule:
# - A soma de a e b;
# - A multiplicação elemento a elemento (a * b);
# - O produto escalar de a e b.


```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

soma = a + b
multiplicacao = a * b
escalar = 0 
for i in range(3):
    produto = a[i] * b[i]
    escalar += produto
print(f"{soma},{multiplicacao} ,{escalar}")
```

# 3.1 Dadas as matrizes A = [[1, 2], [3, 4]] e B = [[5, 6], [7, 8]], calcule:
# - Soma;
# - Produto interno (escalar);
# - Transposta de A.

```python
import numpy as np
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
soma = a + b
escalar = 0 

for i in range(len(a+b)):
    for j in range (len(a+b)):
        produto = a[i][j] * b[i][j]
        escalar += produto
print(f"a soma das matrizes:\n {soma} \n, o produto escalar: {escalar}\n")

transpostaA=a.T
print(transpostaA)
```

# 5.1 Dado o vetor y = [1, 5, 3, 8, 2], calcule:
# - A média, mediana e desvio padrão;
# - O valor máximo e mínimo.

```python
import numpy as np
y = np.array([1,5,3,8,2])

media =  np.mean(y)
print(f"{media}")
mediana = np.median(y)
print(f"{mediana}")
desvio_padrao = np.std(y)
print(f"{desvio_padrao}")
minimo = y.min()
print(minimo)
maximo = y.max()
print(f"{maximo}")
```
