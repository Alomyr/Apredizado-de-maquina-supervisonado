```python
from sklearn.datasets import load_linnerud
import matplotlib.pyplot as plt
import pandas as pd

# from google.colab import drive
# drive.mount('/content/drive/')
# import sys
# sys.path.insert(0, '/content/drive/MyDrive/5 semestre /Aprendizado de maquina supervisionado/IMD3002 - Aprendizado de Máquina Supervisionado/1-unidade')
# from toolbox import dataset_iris

# df_iris = dataset_iris()
# df_iris.head()

load = load_linnerud()
dir(load)
```

```python

df_ex = pd.DataFrame(data=load.data, columns=load.feature_names)
df_ex.head()
```

```python

df_metricas = pd.DataFrame(data=load.target, columns=load.target_names)
df_metricas.head()
```

```python

df = pd.concat([df_ex,df_metricas], axis=1)
df.head()
```

# A média e desvio padrão dos pesos dos candidatos:


```python
print(f"Media dos pesos dos candidatos: {df["Weight"].mean()}")
print(f"Desvio padrao dos pesos dos candidatos: {df["Weight"].std()}")
```

# Calcule o Intervalo Interquartil do números de saltos ("Jumps") dos candidatos:

```python
Q1 = df["Jumps"].quantile(0.25)
Q3 = df["Jumps"].quantile(0.75)
intervalo = Q3-Q1
print(f"Intervalo interquartil: {intervalo}")
```

# Extraia os valores de flexões (Chins) e pulso (Pulse) no formato de listas
# para fazer um scatter plot da relação flexão vs pulso. Não esqueça de dar os 
# nomes aos eixos x e y.



```python
list_chins = df["Chins"].values.tolist()
print(list_chins)
list_pulse = df["Pulse"].values.tolist()
print(list_pulse)

plt.scatter(list_chins, list_pulse, color="purple", marker=0)
plt.title("Grafico entre chins e pulse")
plt.xlabel(list_chins)
plt.ylabel(list_pulse)
plt.show()
```

# Plote um histograma da distribuição do tamanho das cinturas (Waist) dos candidatos:

```python
list_waist = df["Waist"].values.tolist()
plt.hist(list_waist, bins=30, color="purple", alpha=0.5)
plt.title("Histograma da Waist dos candidatos")
plt.xlabel(list_waist)
plt.show()
```
