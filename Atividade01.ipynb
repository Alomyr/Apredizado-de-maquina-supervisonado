# 2.1 Dados os vetores a = [1, 2, 3] e b = [4, 5, 6], calcule:
# - A soma de a e b;
# - A multiplicação elemento a elemento (a * b);
# - O produto escalar de a e b.


```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

soma = a + b
multiplicacao = a * b
escalar = 0 
for i in range(3):
    produto = a[i] * b[i]
    escalar += produto
print(f"{soma},{multiplicacao} ,{escalar}")
```
